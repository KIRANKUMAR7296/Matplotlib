# Import Packages :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Data Set :

Immigration to Canada **IBM** Data Set

OdName : Origin | Destination Name | Country Name

REG : Region

RegName : Region of the Continent.

AreaName : Continent Name

DEV : Development Region

```python
data = pd.read_excel('Canada Immigration.xlsx',sheet_name = 'Canada by Citizenship',skiprows=range(20),skipfooter=2)

data.head()
```

# Data Set Information :

```python
data.info()
```

Conclusion :

Data Set has 195 Rows and 43 Columns 

No Missing Data

# Features with Data Type Object :

```python
data.select_dtypes('object').head()
```

# Unique Value Counts :

```python
print(f'Type : \n{data["Type"].value_counts()}\n')

print(f'Coverage : \n{data["Coverage"].value_counts()}\n')

print(f'Region Name : \n{data["RegName"].value_counts()}\n')

print(f'Development Region Name : \n{data["DevName"].value_counts()}\n')
```

# Drop Unnecessary Features :

```python
data.drop(['AREA','REG','DEV','Type','Coverage'], axis = 1, inplace = True)

data.head()
```

# Rename Features :

* OdName : Name of Country.

* AreaName : Name of Continent.

* RegName : The Region Name of the Continent.

```python
data.rename(columns = {'OdName' : 'Country', 'AreaName' : 'Continent', 'RegName' : 'Region'}, inplace = True)

data.head()
```

Now the Data Set is more Understandable.

We Have the Country Name, Continent Name, Region Name,

The Development Region : Country is a Developing Country or not.

All the Other Features Represents The Year which Consist of Number of Immigrants in that Particular Year.

Creating a Column for **Total Sum** of All Immigrants from 1980 to 2013 from Each Country.

```python
data['Total'] = data.sum(axis = 1) # Sum will Add the values only from Features of Data Type (Numeric)

data.head()
```

# Set Country Feature as Index :

```python
data = data.set_index('Country')

data.head()
```

# Visualizing the Migration Pattern in Canada

```python
# Styles Available in Matpotlib for Plotting :
print(f'Matplotlib Styles : \n{plt.style.available}')
```

```python
# Set Style :
plt.style.use('fivethirtyeight')
```

# 1.Line Plot

Useful to See Country's Immigration to Canada by Year.

Make a List of the Years 1980 to 2013.

```python
years = list(map(int, range(1980, 2014)))

data.loc[['Switzerland'],years].T.head()
```

```python
data.loc['Switzerland',years].plot()
plt.title('Immigration from Switzerland to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years');
```

# Comparing Asian Countries :

Observing the Immigration Trends over the Years for India, Pakistan and Bangladesh.

India and Neighbouring Countries.

```python
data.loc[['India','Pakistan','Bangladesh'],years].T.tail()
```

```python
data.loc[['India','Pakistan','Bangladesh'],years].T.plot();
```

# 2.Pie Plot :

Total Number of Immigrants for Each Continent.

```python
cont = data.groupby('Continent', axis = 0).sum()

cont['Total'].plot(kind = 'pie', figsize = (5,5), autopct = '%1.0f%%', shadow = True)

plt.title('Immigration By Continents')

plt.axis('equal');
```

```python
# Improve Pie Chart :

explode = [0.1,0,0,0,0.1,0.1]

cont['Total'].plot(kind = 'pie', figsize = (15,5), autopct = '%1.0f%%', labels = None, pctdistance = 1.15, explode = explode)

plt.axis('equal')

plt.legend(labels = cont.index, loc='upper right', fontsize = 12);
```

# 3.Box Plot :

```python
data.loc[['China'],years].T.plot(kind = 'box', figsize = (5,5))
plt.title('Box Plot of Chinese Immigrants')
plt.ylabel('Number of Immigrants');
```

```python
data.loc[['India','Pakistan','Bangladesh'],years].T.plot(kind = 'box', figsize = (7,7))
plt.title('Box Plot of India, Pakistan and Bangladesh Immigrants')
plt.ylabel('Number of Immigrants');
```

# 4.Scatter Plot :

Best for Understanding the Relationship between Continuous Values

```python
totalperyear = pd.DataFrame(data[years].sum(axis = 0))

totalperyear.head()
```

```python
totalperyear.index = map(int, totalperyear.index)
totalperyear.reset_index(inplace=True)
totalperyear.columns = ['Year','Total']
totalperyear.head()
```

```python
totalperyear.plot(kind = 'scatter', x = 'Year', y = 'Total', figsize = (10,6), color = 'red')
plt.title('Total Immigration from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants');
```

# 5.Area Plot :

Area Covered Under Line Plot.

```python
top = data.loc[['India','China','Pakistan','Bangladesh','France'], years].T
top.head()
```

```python
top.plot(kind = 'area', stacked = False, figsize = (20,10))
plt.title('Immigration trend from Europe')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years');
```

# 6.Histogram :

Distribution of Variable

```python
data[2012].plot(kind = 'hist', figsize = (8,5))
plt.title('Histogram of Immigration from 195 Countries in 2012')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants');
```

**Stacked Plot** of the **Histogram**

```python
top.plot.hist(figsize = (12,6))
plt.title('Histogram of Immigration from some Populous Countries')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants');
```

Alpha : Determines the **Transparency** of the Color.

For Overlapping Plots, **Transparency** Help us to **Visualize** the Shape of **Each Distribution.**

```python
count, bin_edges = np.histogram(top,15)
top.plot(kind = 'hist', figsize = (12,6), bins = 15, alpha = 0.6,xticks = bin_edges);
```

```python
top.plot(kind = 'hist',figsize = (12,6), bins = 15, xticks = bin_edges, stacked = True)
plt.title('Histogram of Immigration from Some Populous Countries')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants');
```

# 7.Bar Plot :

Number of **Immigrants Per Year.**

```python
data.loc['France', years].plot(kind = 'bar', figsize = (12,6))
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Immigrants from France');
```

The Plot Shows an **Increasing Trend** since 1997 for over a Decade, It could be Worth Mentioning.

It can be done using **Annotation Function.**

```python
data.loc['France', years].plot(kind = 'bar', figsize = (12,6))
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Immigrants from France')
plt.annotate('Increasing Trend', xy = (19,4500), rotation = 18, va = 'bottom', ha = 'left')
plt.annotate('',xy = (29,5500), xytext = (17,3800), xycoords = 'data', arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3', color = 'black', lw = 1.5));
```

**Horizontal Bar Plot** makes it More Understandable, Labels on the Bar can be ever Better.

```python
data.loc['France', years].plot(kind = 'barh', figsize = (12,16))
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Immigrants from France')

for index,value in enumerate(data.loc['France', years]): 
    label = format(int(value),',')
    plt.annotate(label, xy = (value - 300, index - 0.1),color = 'black');
```
